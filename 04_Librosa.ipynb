Bensound - [Badass](https://www.bensound.com/royalty-free-music/track/badass) (Urban/Groove) 

```python
ipd.Audio('bensound-badass.mp3')
```

```python
# Carrega arquivo de áudio
x, sr = librosa.load('bensound-badass.mp3')

# Wave plot
plt.figure(figsize=(30, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
# Extraindo Zero Crossings 
zero_cross = librosa.zero_crossings(x, pad=False)
zero_cross
```

```python
# Total de Zero Crossings 
total_zero_cross = sum(zero_cross)
print(total_zero_cross)
```

```python
# Centroid Spectral 
centroid_spectral = librosa.feature.spectral_centroid(x, sr=sr)
centroid_spectral
```

```python
# Obtém a média do centroid spectral 
mean_centroid_spectral = np.mean(centroid_spectral)
mean_centroid_spectral
```

```python
# Obtém o desvio padrão do centroid spectral 
std_centroid_spectral = np.std(centroid_spectral)
std_centroid_spectral
```
