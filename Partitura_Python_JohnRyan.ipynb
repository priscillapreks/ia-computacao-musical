```python
# Ler partitura
import music21 

partitura = music21.converter.parse('johnryans.abc')
partitura.show()
```

```python
# primeira partitura
partitura = partitura[0]
partitura.show()
```

```python
# Contagem das notas da partitura 
partStream = partitura.parts.stream()

notasMusicais = []

for n in partStream.flat.notes:
  nomeNota = n.pitch.name
  oitavaNota = n.pitch.octave
  notasMusicais.append(str(nomeNota) + '' + str(oitavaNota))
  
print(f'Notas: {notasMusicais}')
```

```python
# Frequência das notas da partitura
import matplotlib.pyplot as plt

plt.xlabel('Nota')
plt.ylabel('Frequência')

plt.hist(notasMusicais, bins=25)
plt.show()
```

```python
# Contagem de notas 

for i in set(notasMusicais):
  count = notasMusicais.count(i)
  print(f'{i}:{count}') 
```
